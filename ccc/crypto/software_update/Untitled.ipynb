```python
cnds = []
for i in range(300):
  fname='fake%03d'%i
  tmp = hashlib.sha256(fname.encode('ASCII') + '\x00')
  print(fname, tmp.hexdigest())
  cnds.append((fname, bytearray(tmp.digest())))
```

```python
def to_vec(x):
  n = len(x)*8
  res  =vector(GF(2), n)
  for i in range(n):
    res[i] = x[i//8]>>(i&7)&1
  return res
    
m = matrix(GF(2), 256, len(cnds))
for i, v in enumerate(cnds):
  m[:,i] = to_vec(v[1])

b = to_vec(target) + to_vec(cur)
x = m.solve_right(b)


print(len(x))
```
